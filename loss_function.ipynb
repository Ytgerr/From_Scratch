### Loss Function

 
–§—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å ‚Äî —ç—Ç–æ –º–µ—Ä–∞ —Ç–æ–≥–æ, –Ω–∞—Å–∫–æ–ª—å–∫–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ $\hat{y}$ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –æ—Ç –∏—Å—Ç–∏–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π $y$.  
–ú–æ–¥–µ–ª—å –æ–±—É—á–∞–µ—Ç—Å—è —Ç–∞–∫, —á—Ç–æ–±—ã **–º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å** –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ.  

–ï—Å–ª–∏ –æ–±–æ–∑–Ω–∞—á–∏—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ –∫–∞–∫ $\langle w, x_i \rangle$, —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å –¥–ª—è –æ–¥–Ω–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –∫–∞–∫:

$$
L(y_i,\langle w, x_i \rangle)
$$

–ê –¥–ª—è –≤—Å–µ–π –≤—ã–±–æ—Ä–∫–∏ (—Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ) ‚Äî –∫–∞–∫:

$$
\text{Loss} = \frac{1}{N} \sum_{i=1}^{N} L(y_i, \langle w, x_i \rangle)
$$

**–î–ª—è —á–µ–≥–æ –Ω—É–∂–Ω—ã —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å?**
1. –û–ø—Ä–µ–¥–µ–ª—è—é—Ç –∫—Ä–∏—Ç–µ—Ä–∏–π –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏.  
2. –ü–æ–∑–≤–æ–ª—è—é—Ç –≤—ã–±–∏—Ä–∞—Ç—å –º–æ–¥–µ–ª—å –∏ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã.  
3. –°–≤—è–∑–∞–Ω—ã —Å –º–µ—Ç—Ä–∏–∫–∞–º–∏ –∫–∞—á–µ—Å—Ç–≤–∞: –∏–Ω–æ–≥–¥–∞ –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏—è loss –Ω–∞–ø—Ä—è–º—É—é —É–ª—É—á—à–∞–µ—Ç –º–µ—Ç—Ä–∏–∫–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, MSE –∏ RMSE), –∏–Ω–æ–≥–¥–∞ –Ω–µ—Ç (–Ω–∞–ø—Ä–∏–º–µ—Ä, log loss –∏ accuracy).  

**–†–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è** —á–∞—Å—Ç–æ –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è –∫–∞–∫ —à—Ç—Ä–∞—Ñ –∫ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å, —á—Ç–æ–±—ã —É–º–µ–Ω—å—à–∏—Ç—å –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ:

$$
\text{Loss}_{\text{reg}} = \text{Loss} + \lambda \cdot R(w)
$$

–≥–¥–µ $R(w)$ ‚Äî —à—Ç—Ä–∞—Ñ (L1 –∏–ª–∏ L2), $\lambda$ ‚Äî –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏.


# –û—Å–Ω–æ–≤–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å

### üîπ –î–ª—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
- **MSE (Mean Squared Error)**  
$$
L(y_i, \hat{y}_i) = (y_i - \hat{y}_i)^2
$$
- **MAE (Mean Absolute Error)**  
$$
L(y_i, \hat{y}_i) = |y_i - \hat{y}_i|
$$
- **Huber Loss** ‚Äî —Å–æ—á–µ—Ç–∞–Ω–∏–µ MSE –∏ MAE, —É—Å—Ç–æ–π—á–∏–≤–µ–µ –∫ –≤—ã–±—Ä–æ—Å–∞–º  
$$
L_\delta(y_i, \hat{y}_i) = 
\begin{cases} 
\frac{1}{2}(y_i - \hat{y}_i)^2 & \text{–µ—Å–ª–∏ } |y_i - \hat{y}_i| \le \delta \\[2mm]
\delta \cdot |y_i - \hat{y}_i| - \frac{1}{2}\delta^2 & \text{–∏–Ω–∞—á–µ}
\end{cases}
$$

### üîπ –î–ª—è –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
- **Log Loss / Binary Cross-Entropy**  
$$
L(y_i, \hat{y}_i) = - \big( y_i \log \hat{y}_i + (1 - y_i) \log (1 - \hat{y}_i) \big)
$$
- **Hinge Loss** (–¥–ª—è SVM)  
$$
L(y_i, \hat{y}_i) = \max(0, 1 - y_i \cdot \hat{y}_i)
$$
- **Exponential Loss** (–¥–ª—è AdaBoost)  
$$
L(y_i, \hat{y}_i) = \exp(-y_i \cdot \hat{y}_i)
$$

### üîπ –î–ª—è –º—É–ª—å—Ç–∏–∫–ª–∞—Å—Å–æ–≤–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
- **Multiclass Cross-Entropy (Softmax Loss)**  
$$
L(y_i, \hat{y}_i) = - \sum_{k=1}^{K} y_{i,k} \log \hat{p}_{i,k}
$$
